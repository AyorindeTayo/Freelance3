**Decision tree optimization Process**

creating a column for ishs*iref

```python
df['ishandiref'] = df['ishs']*df['iref']
```

```python
df.head()
```

Dividing iref/iprd 

```python
df['irefandiprd'] = df['iref']/df['iprd']
```

```python
df.head()
```

Creating the features for the decision tree

```python
list_features=['ishandiref','irefandiprd','pctA','AMPA','PMPA','AggPctSpread','PassPctSpread','pctOxprAcrossSpread','pctDecOSentOrAcrossSpread','SentExecAcrossSpreadDiff']
```

```python
X=df[list_features]
```

```python
X
```

```python
df['Return'] =df['in_net_pnl'].pct_change(80).shift(-80)
y=np.where(df.Return>0,1,0)
```

```python
df
```

```python
y
```

```python
y[0:10]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,stratify=y)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treeClassifier = DecisionTreeClassifier(max_depth=3,min_samples_leaf=6)
treeClassifier.fit(X_train,y_train)
```

```python
y_pred=treeClassifier.predict(X_test)
```

 **The signal is showing that we can sell with precision of 84% and no signal at all for buying**

```python
from sklearn.metrics import classification_report
report= classification_report(y_test,y_pred)
print(report)
```

```python
from sklearn import tree
import graphviz
data=tree.export_graphviz(treeClassifier,filled=True, feature_names=list_features,class_names=np.array(['0','1']))
graphviz.Source(data)
```
